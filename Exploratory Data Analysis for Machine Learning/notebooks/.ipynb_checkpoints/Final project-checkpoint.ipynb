## Brief description of the data set and a summary of its attributes

**Dataset name:**

Key Indicators of Heart Disease

2020 annual CDC survey data of 400k adults related to their health status

**Description:**

According to the CDC, heart disease is one of the leading causes of death for people of most races in the US (African Americans, American Indians and Alaska Natives, and white people). About half of all Americans (47%) have at least 1 of 3 key risk factors for heart disease: high blood pressure, high cholesterol, and smoking. Other key indicator include diabetic status, obesity (high BMI), not getting enough physical activity or drinking too much alcohol. Detecting and preventing the factors that have the greatest impact on heart disease is very important in healthcare. Computational developments, in turn, allow the application of machine learning methods to detect "patterns" from the data that can predict a patient's condition.

**Summary of attributes:**

The dataset contains 18 variables (9 booleans, 5 strings and 4 decimals). In machine learning projects, "HeartDisease" can be used as the explonatory variable, but note that the classes are heavily unbalanced.

- **HeartDisease:** Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI).
- **BMI:** Body Mass Index
- **Smoking:** Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes]
- **AlcoholDrinking:** Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week
- **Stroke:** (Ever told) (you had) a stroke?
- **PhysicalHealth:** Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good? (0-30 days)
- **MentalHealth:** Thinking about your mental health, for how many days during the past 30 days was your mental health not good? (0-30 days)
- **DiffWalking:** Do you have serious difficulty walking or climbing stairs?
- **Sex:** Are you male or female?
- **AgeCategory:** Fourteen-level age category
- **Race:** Imputed race/ethnicity value
- **Diabetic:** (Ever told) (you had) diabetes?
- **PhysicalActivity:** Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
- **GenHealth:** Would you say that in general your health is...
- **SleepTime:** On average, how many hours of sleep do you get in a 24-hour period?
- **Asthma:** (Ever told) (you had) asthma?
- **KidneyDisease:** Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease? 
- **SkinCancer:** (Ever told) (you had) skin cancer?

### Outliers Treatment

```python
heart_2020.describe()
```

For the treatment of the outliers we start by visualizing the boxplots of our numerical variables.

**Full answer from the dataset creator: [https://www.kaggle.com/datasets/kamilpytlak/personal-key-indicators-of-heart-disease/discussion/315289]:**

Considering the following continuous variables:

- **BMI** - min.: 12; max.: 94.8 - with a BMI of 94.8, a person must be extremely morbidly obese (which is not that extremely improbable, given the prevalence of obesity in the US)
- (**PhysicalHealth**, **MentalHealth**) - this is a daily scale, from 0 to 30 days, outliers are hard to find here
- **SleepTime** - we can see above that the mean as well as the median are quite close to each other: the average sleep time of the respondents is between 7 and 8 hours, but we also have strange answers, such as 1 hour or 24 hours, which doesn't seem to be possible. When excluding these values, you can use the interquartile range, exclude 5% of extreme observations or use some statistical tests to determine whether a certain observation is an outlier.

```python
for i in heart_2020.select_dtypes(float):
    plt.figure()
    sns.boxplot(x=heart_2020[i])
```

We show the number of outliers for each of the numerical variables in the dataset

```python
cols = ['SleepTime', 'BMI']

Q1 = heart_2020[cols].quantile(0.25)
Q3 = heart_2020[cols].quantile(0.75)
IQR = Q3 - Q1

print('INITIAL NUMBER OF OUTLIERS:')

((heart_2020[cols] < (Q1 - 1.5 * IQR)) | (heart_2020[cols] > (Q3 + 1.5 * IQR))).sum()
```

The next step is to eliminate the outliers we have detected. We will only eliminate outliers from the **BMI** and **SleepTime** variables since the PhysicalHealth and MentalHealth variables are scale variables ranging from 0 to 30 days.

```python
print('INITIAL SHAPE WITH OUTLIERS:', heart_2020.shape)
heart_2020 = heart_2020[~((heart_2020[cols] < (Q1 - 1.5 * IQR)) | (heart_2020[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print('FINAL SHAPE WITHOUT OUTLIERS:', heart_2020.shape)
```

We visualize again the boxplots of our numerical variables.

As can be seen, most of the outliers have been eliminated, without significantly reducing the sample (we eliminated only **4.5%**).

```python
for i in cols:
    plt.figure()
    sns.boxplot(x=heart_2020[i])
```

Next we will start with the **treatment of the categorical variables**.

First, we show all the different categories of our categorical variables as well as their count.

As can be seen, all categories have significant samples, so there is no need to regroup into other more general categories.

```python
categorical_vars.insert(0,'HeartDisease')
```

```python
binary_categorical_vars = []
ordinal_categorical_vars = []
one_hot_categorical_vars = []
```

```python
for categorical_var in categorical_vars:
    value_counts = heart_2020[categorical_var].value_counts()
    if len(value_counts.index) == 2:
        binary_categorical_vars.append(categorical_var)
    elif len(value_counts.index) > 2 and categorical_var != 'GenHealth':
        one_hot_categorical_vars.append(categorical_var)
    else:
        ordinal_categorical_vars.append(categorical_var)
    print(value_counts)
    print('-'*40)
```

We will use **Binary encoding** for variables that have only two possible labels, **Ordinal encoding** for the GenHealth variable and **One-Hot encoding** for the rest of the variables.

```python
# Binary encoding
encoder = LabelEncoder()
for binary_categorical_var in binary_categorical_vars:
    heart_2020[binary_categorical_var] = encoder.fit_transform(heart_2020[binary_categorical_var])

# One-Hot encoding
heart_2020 = pd.get_dummies(heart_2020, columns=one_hot_categorical_vars, drop_first=True)

# Ordinal encoding
gen_health_map = {'Very good':3,'Fair':1,'Good':2,'Poor':0,'Excellent':4}
heart_2020[ordinal_categorical_vars[0]]=heart_2020.loc[:,ordinal_categorical_vars[0]] = heart_2020[ordinal_categorical_vars[0]].map(gen_health_map)
```

## Hypothesis Testing

### First hypothesis

The first hypothesis consists of a **non-parametric test** on the **BMI** variable, known as the **KOLMOGOROV - SMIRNOV CONTRAST (goodness-of-fit test)**.

**Explanation:** It contrasts whether a data set comes from a population with a certain probability distribution, so its use is similar to the Chi-square contrast, but it works with the distribution function instead of the quantity or density function. It is preferably used when the variable is of a continuous type.

    H0: The BMI variable is distributed as a normal distribution.
    H1: The BMI variable is not distributed as a normal distribution.
    
From the output we can see that the test statistic is 1.0 and the corresponding p-value is 0.0. Since the p-value is less than .05, we reject the null hypothesis. We have sufficient evidence to say that the BMI data does not come from a normal distribution.

```python
#alfa = 0.05

data = np.array(heart_2020['BMI'])
np.random.shuffle(data)

stats.kstest(data, 'norm')
```

### Second hypothesis

The second hypothesis consists of a **non-parametric test** on the **BMI** and **SleepTime** variables, known as the **KOLMOGOROV - SMIRNOV CONTRAST (goodness-of-fit test) for two samples**.

**Explanation:** Performs the two-sample Kolmogorov-Smirnov test for goodness of fit. This test compares the underlying continuous distributions F(x) and G(x) of two independent samples. 

    H0: The BMI and SleepTime variables have identical distributions, F(x)=G(x) for all x.
    H1: The BMI and SleepTime variables don't have identical distributions.
    
From the output we can see that the test statistic is 1.0 and the corresponding p-value is 0.0. Since the p-value is less than .05, we reject the null hypothesis. We have sufficient evidence to say that the BMI and SleepTime don't have identical distribution.

```python
#alfa = 0.05

data1 = np.array(heart_2020['BMI'])
data2 = np.array(heart_2020['SleepTime'])

np.random.shuffle(data1)
np.random.shuffle(data2)

stats.ks_2samp(data1, data2)
```

### Third hypothesis

The third hypothesis consists of a **non-parametric test** on the **HeartDisease** and **Stroke** variables, known as the **PEARSON'S CHI-SQUARE (χ2) CONTRAST (test of independence)**.

**Explanation:** It consists of checking whether two characteristics are related to each other, and allows us to work with all types of variables, qualitative, nominal and ordinal or quantitative (if they are continuous in intervals). 

We can answer questions such as: is the salary of workers related to their seniority, does the number of traffic accidents depend on the age of the driver, does being a smoker depend on the development of lung cancer, do the grades of the students depend on the assigned teacher, etc.).

    H0: The HeartDisease and Stroke variables are independent.
    H1: The HeartDisease and Stroke variables are not independent.

For this type of contrast, a contingency table should be made as a first step. A contingency table (also called a cross-reference table) is used in statistics to summarize the relationship between several categorical variables.

```python
contingency_table = pd.crosstab(index=heart_2020['HeartDisease'], columns=heart_2020['Stroke'])
contingency_table
```

Since,

p-value < alpha 

Therefore, we reject H0, i.e., The HeartDisease and Stroke variables are not independent.

```python
data = contingency_table.values
print(data)
stat, p, dof, expected = stats.chi2_contingency(data)

alpha = 0.05
print("p value is " + str(p)) 
if p <= alpha: 
    print('Dependent(reject H0)') 
else: 
    print('Independent(H0 holds true)')
```
